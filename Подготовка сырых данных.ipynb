```python
import pandas as pd
import numpy as np

import matplotlib
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
billboard = pd.read_csv('билборд.csv', encoding = 'ISO-8859-1')
billboard.head()
```

```python
billboard.columns
```

```python
plt.plot(billboard.loc[0, 'x1st.week': 'x76th.week'])
```

```python
plt.plot(range(1, 77), billboard.loc[0, 'x1st.week': 'x76th.week'])
```

```python
for index, row in billboard.iterrows():
    plt.plot(range(1, 77), row['x1st.week': 'x76th.week'], color = "C0", alpha = 0.1)
```

```python
bshort = billboard[['artist.inverted', 'track', 'time', 'date.entered', 'x1st.week', 'x2nd.week', 'x3rd.week', 'x76th.week']]
bshort.head()
```

```python
bmelt = bshort.melt(['artist.inverted', 'track', 'time', 'date.entered'],['x1st.week', 'x2nd.week', 'x3rd.week'], 'week', 'rank')
bmelt.head()
```
